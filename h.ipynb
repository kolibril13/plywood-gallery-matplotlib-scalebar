```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_scalebar.scalebar import ScaleBar
import matplotlib.image as mpimg
%config InlineBackend.figure_format='retina'
%config InlineBackend.print_figure_kwargs = {'pad_inches':0}

def imshow_bar(im,scalebar):
    fig, ax = plt.subplots(figsize=(2,2), dpi=60)
    ax.axis("off")
    ax.imshow(im)
    ax.add_artist(scalebar);

img = mpimg.imread('orange.png')
```

```python
scalebar = ScaleBar(0.1, "mm", location="upper right")
imshow_bar(img,scalebar) 
```

```python
scalebar = ScaleBar(0.1, "mm", location="upper left")
imshow_bar(img,scalebar) 
```

```python
scalebar = ScaleBar(0.1, "mm", location="lower left")
imshow_bar(img,scalebar) 
```

```python
scalebar = ScaleBar(0.1, "mm", location="lower right")
imshow_bar(img,scalebar) 
```

```python
img = mpimg.imread('yellow.png')
```

```python
scalebar = ScaleBar(0.1, "mm")
imshow_bar(img,scalebar) 
```
